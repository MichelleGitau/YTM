```python
x = data.drop(['YTM', 'Date','Issue No'], axis=1)
x.head()
```

```python
y = data['YTM']
y.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot = OneHotEncoder(sparse=False, drop='first', handle_unknown='error')

enc = onehot.fit_transform(x[['Bondtype']])
df = pd.DataFrame(index=x.index, data=enc, columns=onehot.get_feature_names(['Bondtype']))
x = pd.concat([x, df], axis=1)
x.head()
```

```python
x.head()
```

```python
x = x.drop(['Bondtype'], axis=1)
x.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
y_pred = regressor.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import balanced_accuracy_score

dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)

y_pred_dtr = dtr.predict(X_test)
```

```python
r2_score(y_test,y_pred_dtr)
```

```python
import pickle as pkl
filename = 'model.pkl'
pkl.dump(dtr,open('model.pkl', 'wb'))
```
